```python
import pandas as pd
df_train = pd.read_csv("./train.csv")
df_test = pd.read_csv("./test.csv")
```

```python
df.tail()
```

```python
df_target = df.groupby('target')
```

```python
fold = 1
for target, group in df_target:
    df[df['target'] != target].to_csv('./target_%d/train.csv' % fold, index=False)
    group.to_csv('./target_%d/test.csv' % fold, index=False)
    fold += 1
```

```python
df_stance = df.groupby('stance')
df_split = [None] * 5
for i in range(5):
    df_split[i] = []
for stance, group in df_stance:
    group = group.sample(frac=1).reset_index(drop=True)
    gap = int(len(group))//5
    for i in range(5):
        df_temp = group.iloc[i*gap:(i+1)*gap]
        df_split[i].append(df_temp)

# gap = int(len(df))//5
# for i in range(5):
#     df_temp = df.iloc[i*gap:(i+1)*gap]
#     df_temp.to_csv('split_%s.csv' % (i+1), index=False)
```
